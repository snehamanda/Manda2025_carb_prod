```python
import math

# Initial and final sizes
initial_size = 50  # μm
final_size = 2200  # μm

# Growth rates (expressed as decimals)
growth_rates = [0.005, 0.01, 0.015, 0.02, 0.025, 0.03]

# Function to calculate time in days for each growth rate
def calculate_time(initial_size, final_size, growth_rate):
    return math.log(final_size / initial_size) / math.log(1 + growth_rate)

# Calculate days for each growth rate
days_to_maturity = [calculate_time(initial_size, final_size, rate) for rate in growth_rates]
print(days_to_maturity)
```
